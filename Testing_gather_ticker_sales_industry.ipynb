## Below list is generated code in Testing_edgartools
It find tickers that have 10 years of filing

```python
# turn the tickers into a list
tickers_w_10yr_hist = pd.read_csv('tickers_w_10yr_hist.csv')
#tickers_w_10yr_hist.dtypes()
#tickers_w_10yr_hist['ticker'] = tickers_w_10yr_hist['ticker'].str.strip()
tickers_list = tickers_w_10yr_hist['ticker'].tolist()
```

```python
tickers_list
```

```python
ticker_rev_sector
```

```python
def pull_rev_sector(ticker):
    
    try:
        yf_obj = yf.Ticker(ticker)
        rev = yf_obj.info['totalRevenue']
        sector = yf_obj.info['sector']
        return [ticker, rev, sector]
    except:
        rev = 'NA'
        sector = 'NA'
ticker_rev_sector    
```

```python
pull_rev_sector('CTXR')
```

```python
list_ = []

for ticker in tickers_list:
    output = pull_rev_sector(ticker)
    list_.append(output)
    
    
```

```python
list_
```

```python
ticker_rev_sector = pd.DataFrame(list_, columns=['ticker', 'ttm_revenue', 'sector'])
```

```python
ticker_rev_sector['ttm_revenue'] = pd.to_numeric(ticker_rev_sector['ttm_revenue'],
                                                errors='coerce')
```

```python
ticker_rev_sector['rev_mil'] = ticker_rev_sector['ttm_revenue'] / 1000000
```

```python
ticker_rev_sector
```

```python
# save down results as csv in case lost connect and need to query the data again
ticker_rev_sector.to_csv('tickers_ttmRevenue_sector.csv', index=False)
```
