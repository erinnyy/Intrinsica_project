### [Domentation](https://docs.openbb.co/platform/usage/financial_statements#field-names) from openbb

Financial Statements
The typical financial statements consist of three endpoints:

- Balance Sheet: obb.equity.fundamental.balance()
- Income Statement: obb.equity.fundamental.income()
- Cash Flow Statement: obb.equity.fundamental.cash()

The main parameters are:

- symbol: The company's ticker symbol.
- period: 'annual' or 'quarter'. Default is 'annual'.
- limit: Limit the number of results returned, from the latest. Default is 5. For perspective, 150 will go back to 1985. The amount of historical records varies by provider.

### Extract sector, industry_category and last reported annual revenue for a list of tickers

```python
# read the list of tickers in scope to a list
tickers_w_10yr_hist = pd.read_csv('tickers_w_10yr_hist.csv')
tickers_list = tickers_w_10yr_hist['ticker'].tolist()
```

```python
len(tickers_list) # we have 7480 tickers in scope
```

```python
def pull_rev_sector(ticker):
    
    try:
        revenue = obb.equity.fundamental.income(ticker, limit=1).to_df()['revenue'][0]
        sector = obb.equity.profile(ticker).to_df()['sector'][0]
        industry = obb.equity.profile(ticker).to_df()['industry_category'][0]
        return [ticker, revenue, sector, industry]
    except:
        revenue = 'NA'
        sector = 'NA'
        industry = 'NA'
        return [ticker, revenue, sector, industry]
```

```python
pull_rev_sector('ONEI')
```

```python
obb.equity.fundamental.income('ONEI', limit=1).to_df()['revenue'][0]
```

```python
tickers_list = ['NVDA', 'CTXR']
```

```python
# Record the start time
start_time = time.time()

universe_ = []

for ticker in tickers_list:
    output = pull_rev_sector(ticker)
    universe_.append(output)
    
# Record the end time
end_time = time.time()

# Calculate the elapsed time
elapsed_time = end_time - start_time

print("Elapsed time:", elapsed_time, "seconds")
```

```python
universe_
```

```python
df_universe = pd.DataFrame(universe_, columns=['ticker', 'revenue', 'sector', 'industry'])
```
